### Preprocessing

```python
import pandas as pd
import contractions
import nltk

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

```python
path = r"../data/raw/training.1600000.processed.noemoticon.csv"

data = pd.read_csv(path, encoding="latin1")

df = pd.DataFrame(data)

df.columns = ["Label", "Serial", "Date", "Query", "Twitter_id", "Tweet"]
```

```python
df.head()
```

```python
df["Clean_Text"] = df["Tweet"]

df.head()
```

### lowercase

```python
df["Clean_Text"] = df["Tweet"].fillna("").astype(str).str.lower()

df.head()
```

Label 

```python
df.loc[df["Label"] == 4, "Label"] = 1
```

```python
df.tail()
```

### URL removal

```python
pattern = r"https?://\S+"

df["Clean_Text"] = (df["Clean_Text"].fillna("").astype(str).
                    str.replace(pattern, "", regex=True))
```

### mention removal

```python
pattern = r"@\w+"

df["Clean_Text"] = (df["Clean_Text"].fillna("").astype(str).
                    str.replace(pattern, "", regex=True))
```

### hashtag cleaning

```python
pattern = r"#"

df["Clean_Text"] = (df["Clean_Text"].fillna("").astype(str).
                    str.replace(pattern, "", regex=True))
```

### Punctuation removal

```python
pattern = r"[!?_]"

df["Clean_Text"] = (df["Clean_Text"].fillna("").astype(str).
                    str.replace(pattern, "", regex=True))
```

### Space cleaning

```python

df["Clean_Text"] = (df["Clean_Text"].fillna("").astype(str).
                    apply(lambda x: " ".join(x.split())))
```

### contractions

```python
df["Clean_Text"] = df["Clean_Text"].fillna("").astype(str).apply(
    lambda x: (contractions.fix(x))
)
```

```python
df.head()
```

### Remove apostrophe 

```python
pattern = r"'s\b"

df["Clean_Text"] = df["Clean_Text"].astype(str).str.replace(pattern, "", regex=True)
```

### Tokenization

```python
df["Tokens"] = df["Clean_Text"].fillna("").astype(str).apply(word_tokenize)

df.head()
```

### Stopwords cleaning

```python
nltk.download("stopwords")
```

```python
stop_words = set(stopwords.words("english"))

stop_words.discard("not")

df["Tokens"] = df["Tokens"].apply(
    lambda x: [word for word in x if word.lower() not in stop_words]
)

df["Clean_Text"] = df["Tokens"].apply(lambda x: " ".join(x))
```
