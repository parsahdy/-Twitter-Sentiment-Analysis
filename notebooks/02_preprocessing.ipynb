### Preprocessing

```python
import pandas as pd
import contractions
import spacy

from nltk.util import ngrams
from nltk.tokenize import word_tokenize
```

```python
path = r"../data/raw/training.1600000.processed.noemoticon.csv"

data = pd.read_csv(path, encoding="latin1")

df = pd.DataFrame(data)

df.columns = ["Label", "Serial", "Date", "Query", "Twitter_id", "Tweet"]
```

```python
df.head()
```

```python
df["Clean_Text"] = df["Tweet"]

df.head()
```

### lowercase

```python
df["Clean_Text"] = df["Tweet"].fillna("").astype(str).str.lower()

df.head()
```

### URL removal

```python
pattern = r"https?://\S+"

df["Clean_Text"] = (df["Clean_Text"].fillna("").astype(str).
                    str.replace(pattern, "", regex=True))
```

### mention removal

```python
pattern = r"@\w+"

df["Clean_Text"] = (df["Clean_Text"].fillna("").astype(str).
                    str.replace(pattern, "", regex=True))
```

### hashtag cleaning

```python
pattern = r"#"

df["Clean_Text"] = (df["Clean_Text"].fillna("").astype(str).
                    str.replace(pattern, "", regex=True))
```

### Punctuation removal

```python
pattern = r"[!?_]"

df["Clean_Text"] = (df["Clean_Text"].fillna("").astype(str).
                    str.replace(pattern, "", regex=True))
```

### Space cleaning

```python

df["Clean_Text"] = (df["Clean_Text"].fillna("").astype(str).
                    apply(lambda x: " ".join(x.split())))
```

### contractions

```python
big_text = " ".join(df["Clean_Text"])

expanded_words = []

for word in big_text.split():
    expanded_words.append(contractions.fix(word))


big_text = " ".join(expanded_words)
```

```python
tokens = word_tokenize(big_text)
```
